# Simulaci√≥n

- Elegir el dominio de la variable aleatroria.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function_integral(x):
    return x**2
a = 2
b = 5

N = 1000
x = np.random.uniform(a,b,N)
y = np.random.uniform(0, function_integral(b),N)
```

```python
n_bajo_curva = np.sum(y < function_integral(x))
n_bajo_curva
```

```python
respuesta = (b-a) * function_integral(b) * (n_bajo_curva / N)
respuesta                                            
```

```python
plt.plot(x,y,'.')
plt.plot(np.linspace(a,b,100), function_integral(np.linspace(a,b,100)), 'r')
plt.show()
```

**Simulaci√≥n de MonteCarlo pi**

```python
r = 1
N = 100000
x = np.random.uniform(-r,r,N)
y = np.random.uniform(-r,r,N)
circulo = x**2 + y**2 <= r**2
n_dentro = np.sum(circulo)
pi = (n_dentro / N) * 4
pi
```

```python
#Grafica el circulo y el cuadrado
plt.figure(figsize=(6,6))
plt.plot(x[circulo], y[circulo], 'b.', label='Dentro del c√≠rculo')
plt.plot(x[~circulo], y[~circulo], 'r.', label='Fuera del c√≠rculo')
plt.xlim(-r, r)
plt.ylim(-r, r)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Estimaci√≥n de œÄ usando el m√©todo de Monte Carlo')
plt.legend()
```
