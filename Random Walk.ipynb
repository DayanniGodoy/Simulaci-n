```python
import numpy as np
import matplotlib.pyplot as plt



n_steps =200
steps = np.random.choice([-1,1], size=n_steps)
price =100 + np.cumsum(steps)

plt.figure(figsize=(10,5))
plt.plot(price)
plt.title("Random Walk de un precio")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

n_sim = 50000
n_steps = 200

final_positions = []

for _ in range(n_sim):
    steps = np.random.choice([-1,1], size=n_steps)
    final_positions.append(np.sum(steps))

plt.hist(final_positions, bins=80)
plt.title("Distribucin de posiciones finales")
plt.show()
```

```python
import numpy as np

np.random.seed(0)

n_sim =2000
times = [10,50,100,200,400]

for t in times:
    finals = []
    for _ in range(n_sim):
        steps = np.random.choice([-1,1], size=t)
        finals.append(np.sum(steps))
    print(f"t={t}, var{np.var(finals):.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

T =1
n =1000
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W = np.cumsum(increments)

plt.plot(W)
plt.title("Aproximacin a movimiento browniano")
plt.show()
```

Calculamos el promedio (valor esperado) de esas 200 caminatas aleatorias.
En este caso, esas caminatas son precios.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
n_steps =200
n_walks =200
all_walks = []
for _ in range(n_walks):
    steps = np.random.choice([-1,1], size=n_steps)
    walk =100 + np.cumsum(steps)
    all_walks.append(walk)
    plt.plot(walk, color='lightgray', alpha=0.5)
average_walk = np.mean(all_walks, axis=0)
plt.plot(average_walk, color='blue', label='Promedio de Caminatas', linewidth=2)
plt.title("200 Caminatas Aleatorias y su Promedio")
plt.xlabel("Tiempo")
plt.ylabel("Walk")
plt.legend()
plt.show()
```
